# Dependencies/Imports

```python
import pandas as pd
from IPython.display import HTML
```

```python
data_path = 'Resources/cities.csv'

cities_df = pd.read_csv(data_path)

# setting all cities to uppercase
cities_df["City"] = cities_df["City"].str.upper()
```

# Organizing Data

```python
# organizing the df before exporting
city_climate_df = cities_df[
    ['City',
     'Lat',
     'Lng',
     'Country',
     'Max Temp',
     'Wind Speed',
     'Cloudiness',
     'Humidity',
    ]]

city_climate_df.head()
```

# Creating HTML Table

```python
cities_html = cities_df.to_html(classes='table table-striped table-hover')

cities_html
```

```python
with open('cities.html', 'w') as fo:
    fo.write(cities_html)
```
